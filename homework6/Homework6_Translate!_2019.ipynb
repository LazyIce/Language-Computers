Now, let's make a very rudimentary MT system!

```python
punct = ".,'()[]:-—"


def mt_v1(spanish_sentence):
    eng_sentence = ""
    for word in spanish_sentence.split(" "):
        word = word.strip(punct)
        if word in span_eng_dict.keys():
            eng_sentence += " " + span_eng_dict[word][0][0]
        elif word.lower() in span_eng_dict.keys():
            eng_sentence += " " + span_eng_dict[word.lower()][0][0]
        else:
            eng_sentence += " " + word
    return eng_sentence


for pair in spanish_english_dev:
    print("v1: " + mt_v1(pair[0]))
    print("GOLD: " + pair[1])

    print("#######")
    print("   ")
```

### Your answer to Question 1

For the work mechanism of "mt_v1", it will split the sentence into words by space and remove the punctuations. Then, it will find the English words for Spanish words in the span_eng_dict. If it can't find the english word for spanish word in dictionary, it will keep the spanish word untranslated.

It removes the punctuaions by strip() function.

It transforms the words to lowercase for capitalization.

It only chooses the first translation in the all possible English translations.

It cannot solve the situation that multiple Spanish words need to be translated as a phrase. For this problem, maybe we can support translate multiwords by using ngrams instead of tranlating words one by one.
Also, it cannot solve the situation that the order of words in a phrase maybe different between Spanish and English. For this problem, maybe we can support grammer checking. 

### Keep reading....

Hopefully, you noticed that "mt_v1" cannot deal with situations where MULTIPLE Spanish words need to be translated as a phrase, such as "de un día para otro" = "overnight".

In order to fix this, here is "mt_v2"...



```python

def fix_ngrams(string, n):
    new_sentence = []
    i = 0
    sentence = string.split(" ")
    while i < len(sentence):
        found_ngram = False
        new_word = sentence[i]
        for j in range(2,n+1):
            if j+i <= len(sentence):
                    ngram = ' '.join(sentence[i:i+j]).strip(punct)
                    ngram_lower = ngram.lower()
                    if ngram in span_eng_dict.keys():
                        new_word = span_eng_dict[ngram][0][0]
                        i += j
                        found_ngram = True
                        print("***matched n-gram!"  + " " + ngram + " == " + span_eng_dict[ngram][0][0])
                    elif ngram_lower in span_eng_dict.keys():
                        new_word = span_eng_dict[ngram_lower][0][0]
                        i += j
                        found_ngram = True
                        print("***matched n-gram!" + " " + ngram + " == " + span_eng_dict[ngram][0][0])
        if not found_ngram:
            i += 1
        new_sentence.append(new_word)
    return " ".join(new_sentence)


    
def mt_v2(spanish_sentence):
    eng_sentence = ""
    ngram_fixed_spanish_sentence = fix_ngrams(spanish_sentence, 5)
    for word in ngram_fixed_spanish_sentence.split(" "):
        word = word.strip(punct)
        if word in span_eng_dict.keys():
            eng_sentence += " " + span_eng_dict[word][0][0]
        elif word.lower() in span_eng_dict.keys():
            eng_sentence += " " + span_eng_dict[word.lower()][0][0]
        else:
            eng_sentence += " " + word
    return eng_sentence


for pair in spanish_english_dev:
    print("v1: " + mt_v1(pair[0]))
    print("v2: " + mt_v2(pair[0]))
    print("GOLD: " + pair[1])

    print("#######")
    print("   ")
```

### Your answer to Question 2

In "mt_v2", it will first find the spanish ngram from 2-gram to 5-gram and replace them with the English translation. Then it will translate the single Spanish words to English words. It doesn't translate words one by one any more like "mt_v1".

"mt_v2" tries to take the situation that in two languages, the same meaning may be expressed by several words in one language and be expressed by one word in the other language into account.

```python
def change_gender(spanish_word):
        new_gender = spanish_word
        if spanish_word[-1:] == 'o':
            new_gender = spanish_word[:-1] + 'a'
        elif spanish_word[-1:] == 'a':
            new_gender = spanish_word[:-1] + 'o'
        return new_gender

def mt_v3(spanish_sentence):
    eng_sentence = ""
    ngram_fixed_spanish_sentence = fix_ngrams(spanish_sentence, 5)
    for word in ngram_fixed_spanish_sentence.split(" "):
        word = word.strip(punct)
        if word in span_eng_dict.keys():
            eng_sentence += " " + span_eng_dict[word][0][0]
        elif word.lower() in span_eng_dict.keys():
            eng_sentence += " " + span_eng_dict[word.lower()][0][0]
        elif change_gender(word) in span_eng_dict.keys():
            eng_sentence += " " + span_eng_dict[change_gender(word)][0][0]
            print("***gender-swapped word in dict! " + word)

        elif change_gender(word.lower()) in span_eng_dict.keys():
            eng_sentence += " " + span_eng_dict[change_gender(word.lower())][0][0]
            print("***gender-swapped word in dict! " + word.lower())
        
        else:
            eng_sentence += " " + word
    return eng_sentence


for pair in spanish_english_dev:
    print("v3: " + mt_v3(pair[0]))
    print("GOLD: " + pair[1])

    print("#######")
    print("   ")
```

### Your answer to Question 4

In "mt_v3", it will translate the original word first. If there is no match, the Spanish word ends with 'o' will be transformed to ending with 'a' and the Spanish word ends with 'a' will be transformed to ending with 'o'. Then it will try to find the English word of this new Spanish word in the dictionary for translation.

"mt_v3" takes the "masculine o, feminine a" rule, i.e. gender rule for words in Spanish into account.

```python

import re
import  nltk
from nltk.corpus import  brown
import copy
from collections import defaultdict
import math

class TrigramLanguageModel:
 
  def __init__(self):
    """Initialize your data structures in the constructor."""
    self.trigramCounts = defaultdict(lambda: 0)
    self.bigramCounts = defaultdict(lambda: 0)
    self.unigramCounts = defaultdict(lambda: 1)  # Laplace smoothing
    self.vocab_size = 1 #  1 because of Laplace smoothing
    self.corpus_size = 0
    self.train()
 
  def train(self):
    """ Takes a corpus and trains the language model. 
    """
    self.unigramCounts["UNK"] = 1  # all unknown words are mapped to a single word UNK
     
    alphanumeric = re.compile(r"\w+")  # want to only look at words, not punctuation
    #                               (in this corpus, punctuation is separated from words with a space)
    counter = 0
    for sent in brown.sents():
        sent = "<start> <start> " + " ".join(sent) + " <end> <end>"
        words_in_line = []
        for word in sent.split(" "):
            if alphanumeric.findall(word):
                words_in_line.append(word)
        if len(words_in_line) > 1:
            for i in range(0, len(words_in_line) - 2): 
                word = words_in_line[i].lower()
                next_word = words_in_line[i+1].lower()
                next_next_word = words_in_line[i+2].lower() 
                
                if word != '<end>' and word != "<start>":
                    self.corpus_size += 1
                     
                    if self.unigramCounts[word] == 1:
                        self.vocab_size += 1 
                        # add one to vocab if you have not seen word before
                     
                    self.unigramCounts[word] += 1     
                    self.bigramCounts[word + "__" + next_word] += 1
                    self.trigramCounts[word + "__" + next_word + "__" + next_next_word] += 1
       
     
 
  def score(self, sentence): 
    """ Takes a string as argument and returns the log-probability of the 
        sentence using the language model, using the data computed in train().
    """
    score = 0.0
    words_in_line = sentence.split(" ")
    for i in range(0,len(words_in_line)-2):
        word = words_in_line[i].lower()
        next_word = words_in_line[i+1].lower()
        next_next_word = words_in_line[i+2].lower()
        if (word != '<end>' and word != "<start>"):
         
            if not (word in self.unigramCounts):  # treat all unknown words as a single word "unk"
                word = "UNK"
             
            if self.trigramCounts[word + "__" + next_word + "__" + next_next_word] >= 1:
                bigram_count = self.bigramCounts[word + "__" + next_word]
                trigram_count = self.trigramCounts[word + "__" + next_word + "__" + next_next_word]
                score += math.log(trigram_count)
                score -= math.log(bigram_count)
         
            elif self.bigramCounts[word + "__" + next_word] >= 1:
                unigram_count = self.unigramCounts[word]
                bigram_count = self.bigramCounts[word + "__" + next_word]
                score += math.log(bigram_count) * 0.30 # penalize it for moving down to bigrams
                score -= math.log(unigram_count)
           
            else:
                count = self.unigramCounts[word]
                score += math.log(count) * 0.30 * 0.30  # penalize it for moving down to unigrams
                score -= math.log(self.corpus_size + self.vocab_size)  
    return score




def choose_best_translation(list_of_lists_of_choices): 
        # takes a list of lists; there is one list for each word;
        #each list is a list of possible translations for that word
        ## takes a LIST of LISTS of translations
        # returns the BEST LIST of words
        language_model = TrigramLanguageModel()
        best_sentence = []
        for i in range(len(list_of_lists_of_choices)):
            choices = list_of_lists_of_choices[i]
            if len(choices) == 1:
                best_choice = choices[0] # if there's only one possible translation, use that one
                best_sentence.append(best_choice)
            else:
                best_score = -1000000000
                best_choice = "XX"
                for choice in choices:
                    if choice == "<null>":
                        choice = ""  
                        # some items in our dictionary can be translated as "<null>" meaning
                            #we just don't translate them into English; e.g. "de" (of) is sometimes not translated to English
                     
                     
                    possible_best_sentence = copy.copy(best_sentence)
                    possible_best_sentence.append(choice)
                    
                    rest_of_sentence = []
                    for j in range(len(list_of_lists_of_choices)):
                        if j > i:
                            rest_of_sentence.append(list_of_lists_of_choices[j][0])  
                            ## just adds first choice for rest of words
                    if rest_of_sentence and possible_best_sentence:
                        possible_best_sentence + rest_of_sentence
                    
                    scorable_sentence = " ".join(possible_best_sentence) 
        
                    score = language_model.score("<start> <start> " + scorable_sentence + " <end> <end>")
                    #print choice + str(score)
                    if score >= best_score:
                        best_score = score
                        best_choice = choice
 
                best_sentence.append(best_choice)
        
        return " ".join(best_sentence)
    
print(choose_best_translation([["She"],["are", "is"],["a writer", "writer", "writers"]]))
print(choose_best_translation([["Its", "it"],["were", "was"],["raining", "rain"]]))

```

```python
from functools import reduce
import operator

punct = ".,'()[]:-—"

def count_ngram(candidate, reference, n):
    ref_dict = {}
    words = reference.split()
    words = [w.strip(punct) for w in words]
    ref_length = len(words)
    limits = len(words) - n + 1
    for i in range(limits):
        ngram = ' '.join(words[i:i+n]).lower()
        if ngram in ref_dict.keys():
            ref_dict[ngram] += 1
        else:
            ref_dict[ngram] = 1

    cand_dict = {}
    words = candidate.split()
    limits = len(words) - n + 1
    for i in range(limits):
        ngram = ' '.join(words[i:i+n]).lower()
        if ngram in cand_dict:
            cand_dict[ngram] += 1
        else:
            cand_dict[ngram] = 1
                
    clipped_count = clip_count(cand_dict, ref_dict)
    count = limits
    
    if clipped_count == 0:
        pr = 0
    else:
        pr = float(clipped_count) / count
    
    c = len(words)
    r = ref_length
    bp = brevity_penalty(c, r)
    return pr, bp

def clip_count(cand_dict, ref_dict):
    count = 0
    for word in cand_dict.keys():
        word_count = cand_dict[word]
        word_max = 0
        if word in ref_dict:
            word_max = max(word_max, ref_dict[word])
        word_count = min(word_count, word_max)
        count += word_count
    return count

def brevity_penalty(c, r):
    if c > r:
        bp = 1
    else:
        bp = math.exp(1-(float(r)/c))
    
    return bp

def geometric_mean(precisions):
    return (reduce(operator.mul, precisions)) ** (1.0 / len(precisions))

def calculate_bleu(candidate_translation, reference_translation):
    precisions = []
    for i in range(2):
        pr, bp = count_ngram(candidate_translation, reference_translation, i+1)
        precisions.append(pr)
    bleu = geometric_mean(precisions) * bp
    return bleu
```

## Your answer to Question 13a 

Barnes would say that whether poussa implies more vigour than the English ‘grew’: hence 4)’s ‘throve’ and 5)’s addition of the intensifying ‘up’. However, google translation uses the word "pushed". Also, Barnes would consider that whether this translation can keep the balance of ‘de fortes mains, de belles couleurs’. The google translation doesn't keep the balance. Besides, Barnes would ask what to do about belles couleurs. All five translators agree that there is no way of preserving the plural form. However, google translation keeps the plural form. The google translation seems to stick closer to the original sentence structure and are less ‘interpretative’.

## Your answer to Question 13b 

For the original French sentence, it is a calm, balanced sentence, in three parts, with a triple alliteration, the part containing the second and third ‘p’ making up the central phrase. The choice for petitesse lies normally between ‘paltriness’ and ‘pettiness’, neither of which is perfect, as they have a slightly more disapproving tinge than petitesse. For google translation, instead of "pettiness" or 'paltriness', it uses "smallness", which loses the meaning. Also, the two words 'now', 'knew', both a wail of assonance and a stuttering of rhythm far from the original.

## Your answer to Question 13c

I prefer to read the Lydia Davis version. Davis’s Madame Bovary is a linguistically careful version, in the modern style, rendered into an unobtrusively American English. It conveys the precision of Flaubert’s prose in this novel, while its syntactical mirroring of the French sometimes brings us closer to Flaubert. Also, Lydia Davis’s version uses an act of the imagination as well as a technician’s proficiency, which provides a more acceptable version. Besides, Davis is profoundly out of sympathy, which can provide us a more objective translation version.
